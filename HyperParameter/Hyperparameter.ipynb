```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('sample_data/diabetes.csv')
```

```python
df
```

```python
df.corr()['Outcome']
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(64,activation='relu',input_dim=8))

model.add(Dense(32,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,batch_size=32,epochs=100,validation_data=(X_test,y_test))
```

```python
def build_model(hp):

  model=Sequential()
  model.add(Dense(64,activation='relu',input_dim=8))

  model.add(Dense(32,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer=hp.Choice('optimizer',values=['adam','sgd','rmspro','adadelta']),loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
tuner =kt.RandomSearch(build_model,objective='val_accuracy',max_trials=5)
```

```python
tuner.search(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
from kerastuner.tuners import RandomSearch
from tensorflow import keras
from tensorflow.keras import layers

# Define the model building function
def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Dense(units=hp.Int('units', min_value=32, max_value=512, step=32),
                           activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# Use Keras Tuner to search for best hyperparameters
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=1,
    directory='my_dir',
    project_name='intro')

# Call search to perform hyperparameter tuning
tuner.search(X_train, y_train,
             epochs=10,
             validation_data=(X_test, y_test))

# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]

# Now fit the best model
best_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
#how to check total neuron

def build_model(hp):
    model = keras.Sequential()

    units=hp.Int('units', min_value=32, max_value=512, step=16)

    model.add(Dense(units=units,activation='relu',input_dim=8))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
```
